```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("customer1.txt",header=None)
df.columns=['id','fname','lname','age','prof','loc']
df
```

```python
df1=df['loc'].value_counts()
df1
```

```python
plt.bar(df1.index,df1.values) # we take x=index of the one we need y= values
plt.xlabel('Location',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Location count Data',color='r',fontsize='20')
```

```python
plt.pie(df1.values,labels=df1.index,autopct='%1.0f%%')
plt.legend(loc='upper left')
plt.tight_layout()
plt.title('Location Percentage',color='g',fontsize=20)
```
