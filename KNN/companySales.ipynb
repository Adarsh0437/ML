```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales_data.csv")
df
```

```python
x=df.iloc[:,0]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#1.
plt.plot(x,y)
```

```python
#2.
plt.plot(x,y,linestyle='dotted',color='r',marker='o',mfc='r',linewidth=3,label='Profit Data')
plt.xlabel("Month Number")
plt.ylabel("Sold Units Number")
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc="lower right")
```

```python
#3 
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]

for i in products:
 plt.plot(df["month_number"],df[i],marker="s",color="g")
plt.xlabel("Month Number")
plt.ylabel("Product sales")
plt.xticks(x)
plt.title("Multline Plot for all product sales")
```

```python
#4
x=df.iloc[:,0]
y=df.iloc[:,3]
plt.scatter(x,y)
```

```python
#5
x=df['month_number']
facecream=df['facecream']
facewash=df['facewash']
plt.bar([a-0.25 for a in x],facecream,width=0.25,align='edge',label='Face Cream Sales')
plt.bar([a+0.25 for a in x],facewash,width=-0.25,align='edge',label='Face Wash Sales')
plt.grid()
plt.xlabel("Month number")
plt.ylabel("Profit")
plt.title('Facecream And Facewash Sales')
plt.legend(loc='upper right')
plt.tight_layout()
plt.xticks=(x)

```

```python
#6
plt.hist(df["total_profit"])
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Histogram of Total Profits")
```

```python
#7
total_sales=df[products].sum()
plt.pie(total_sales,autopct="%1.0f%%",labels=products)
plt.legend(loc="upper left")
plt.tight_layout()
plt.title("Total Sales data for last year of each product")
```

```python
#8
plt.subplot(1,2,1)#1st==> row count,2nd==>column 3rd==>position 
plt.plot(df["month_number"],df["bathingsoap"])
plt.xlabel("Month number")
plt.ylabel("Bathing soap sales units")
plt.title("Graph 1")


plt.subplot(1,2,2)
plt.plot(df["month_number"],df["facewash"])
plt.xlabel("Month number")
plt.ylabel("Facewash sales unit")
plt.title("Graph 2")
plt.suptitle("Bathingsoap and Facewash sales for all months")
plt.tight_layout()
```
