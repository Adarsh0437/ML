```python
#Height and Weight
height=[100,145,140,150]
weight=[45,50,40,52]
plt.plot(height,weight)
plt.xlabel('Height in cm',color='r',fontsize=10)
plt.ylabel('Weight in cm',color='g',fontsize=10)
plt.title('Height-Weight Graph',color='y',fontsize=15)
```

```python
vel=[100,120,140,150,200]
displace=[70,40,30,50,70]
plt.plot(vel,displace)
plt.xlabel('Velocity',color='g',fontsize=15)
plt.ylabel('Displacement',color='b',fontsize=15)
plt.title('Velocity and Displacement Graph',color='y',fontsize=20)
# plt.grid() #grid we can also define which axis we need
plt.grid(axis='y',linestyle='dotted',linewidth=4,color='r')
```

```python
#Scatterplot
x1=[1,4,5,7,8,9,10]
y1=[5,4,7,8,11,2,6]
x2=[3,5,7,8,10,12,3]
y2=[4,7,3,6,9,3,11]
plt.scatter(x1,y1,color='g')
plt.scatter(x2,y2,color='c')
```

```python
#subplot
#partition of a graph into 2 parts to use first part as graph 1 and second part as graph 2
x1=[1,4,5,7,8]
y1=[5,4,7,8,11]
plt.subplot(1,2,1) #1st=row 2nd=Columns and 3rd=position
plt.plot(x1,y1)
plt.xlabel('X-Axis',color='c')
plt.ylabel('Y-Axis',color='m')
plt.title('Graph 1',color='g')
x2=[3,5,7,8,10]
y2=[4,7,3,6,9]
plt.subplot(1,2,2) #1st=row 2nd=Columns and 3rd=position
plt.plot(x2,y2)
plt.xlabel('X-Axis',color='c')
plt.ylabel('Y-Axis',color='m')
plt.title('Graph 2',color='b')
plt.suptitle('Graph 1-Graph 2',color='r') #Super Title Main title for both graph
```

```python
#bar chart
x=[5,7,8,9,10]
y1=[3,7,4,8,3]
plt.bar(x1,y1,color='c')
plt.xlabel('X-Axis',color='c')
plt.ylabel('Y-Axis',color='m')
```

```python
country=['India','Aus','China','Pak','Canada']
count=[3,4,2,1,5]
plt.barh(country,count,color='c') #'bar' for bar graph and 'barh' for bar graph Horizontal
#bar chart 2 axis is needed
```

```python
#Histogram
#No. of frequency observation == count
#Only need one axis
x=[1,1,2,2,2,4,4,4,4,5,5,5,6,6,6,6,6,3,3,3,3,3,3]
plt.hist(x,color='c')
```

```python
#Pie chart
x1=[10,20,25,45]
lab=['Apple','Orange','Pineapple','Mango']
color=['red','orange','yellow','green']
plt.pie(x1,labels=lab,colors=color,autopct='%1.0f%%') #labels passing
plt.legend(loc='lower right')
plt.tight_layout()
```
